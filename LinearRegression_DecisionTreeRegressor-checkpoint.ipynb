## ARX UTILITY

```python
def start_shift(na,nb,nk):

	# riceve in ingresso il termine na e i due vettori nb e nk (vanno passati come vettori anche se hanno 1 elemento solo).
	# riceve inoltre due vettori y e u, che sono i dati che abbiamo relativi rispettivamente all'input e all'output.
	# resituisce due elementi numpy: la matrice M e il vettore teta
	# procediamo subito a convertire tutti i vettori in array di numpy
	min_u=0
	min_a=0

	if na>0:
		inizio_ar=np.array(range(0,-na,-1))-1
		min_a=np.min(inizio_ar)
	else:
		inizio_ar=[]

	if type(nb) is int:
		nb=np.append(nb,np.nan)
		nk=np.append(nk,np.nan)
		num_ex=1
	else:
		num_ex=len(nb)

	if len(nb)==0 or np.nanmax(nb)<=0:
		inizio_ex=[]
	else:
		inizio_ex=np.full([num_ex,int(np.nanmax(nb))],np.nan)
		for i in range(0,num_ex):
			max_ord=int(nb[i])
			max_del=int(nk[i])
			inizio_ex[i,0:max_ord]=np.array(range(0-max_del,-max_ord-max_del,-1))
		min_u=np.nanmin(inizio_ex)

	minimo=np.min([min_a,min_u])
	if len(inizio_ar):
		inizio_ar_traslato=inizio_ar-minimo
	else:
		inizio_ar_traslato=[]

	if len(inizio_ex):
		inizio_ex_traslato=inizio_ex-minimo
	else:
		inizio_ex_traslato=[]

	inizio_out=0
	inizio_out_traslato=-minimo

	return(inizio_out,inizio_ar,inizio_ex,inizio_out_traslato,inizio_ar_traslato,inizio_ex_traslato)

def clean (y,X):

	S=np.sum(X,axis=1)
	bad=np.isnan(S)
	X=np.delete(X,bad,axis=0)
	y=np.delete(y,bad)
	bad=np.isnan(y)
	X=np.delete(X,bad,axis=0)
	y=np.delete(y,bad)
	return (y,X)

def dataset_shift (y,U,na,nb,nk):
    
    inizio_out,inizio_ar,inizio_ex,inizio_out_traslato,inizio_ar_traslato,inizio_ex_traslato=start_shift(na,nb,nk)
    yout=y[int(inizio_out_traslato):]
    L=len(yout)
    X=np.zeros([L,na+np.sum(nb)])
    cont=0
    if len(inizio_ex_traslato):
        for i in range(len(nb)):
            for j in range(nb[i]):
                X[:,cont]=U[int(inizio_ex_traslato[i,j]):int(inizio_ex_traslato[i,j])+L,i].flatten()
                cont=cont+1

    if len(inizio_ar_traslato):
        for j in range(na):
            X[:,cont]=y[int(inizio_ar_traslato[j]):int(inizio_ar_traslato[j]+L)]
            cont=cont+1

    X=np.array(X)
    return(yout,X)

def valid_perf(y_true,y_pred):
	import sklearn.metrics as mt
	r2=mt.r2_score(y_true, y_pred)
	corrcoef=np.corrcoef(y_true,y_pred)
	me=np.average(y_true-y_pred)
	nme=me/np.average(y_true)
	mae=mt.mean_absolute_error(y_true, y_pred)
	mse=mt.mean_squared_error(y_true, y_pred)
	msle=mt.mean_squared_log_error(y_true, y_pred)
	mape=mt.mean_absolute_percentage_error(y_true, y_pred)
	nmae=mae/np.average((np.abs(y_true)))
	medae=mt.median_absolute_error(y_true, y_pred)
	maxe=mt.max_error(y_true, y_pred)
	evar=mt.explained_variance_score(y_true, y_pred)
	perf_dict={}
	for variable in ["corrcoef","r2","me","nme","mae","mse","msle","mape","nmae","medae","maxe","evar"]:
		perf_dict[variable] = eval(variable)

	return(perf_dict)

def nanpredict(model,X):
	L=np.size(X,axis=0)
	#print(L)
	y=np.zeros([L])
	for i in range(0,L):
		S=np.sum(X[i,:])
		bad=np.isnan(S)

		if bad:
			y[i]=np.nan
		else:
			input=np.reshape(X[i,:],[1,-1])
			y[i]=model.predict(input)

	return y

```

## Lettura iniziale dei dati
Estraggo i dati di interesse e li suddivido nei corrispettivi valori

```python
import scipy.io as sio
import numpy as np

#dati dal 2014 al 2019 incluso
all_data = sio.loadmat('dati/dati_tutti_corretti.mat')

#parte autoregressiva
conc_max = all_data['conc_massima']
conc_media = all_data['conc_media']

#parte esogena
temp = [item for sublist in zip(*all_data['T']) for item in sublist]
vel_vento = [item for sublist in zip(*all_data['VV']) for item in sublist]
rad_solare = [item for sublist in zip(*all_data['Rad']) for item in sublist]
emissioni = [item for sublist in zip(*all_data['emi']) for item in sublist]
```

## Creazione delle variabili per il modello

### Variabili per l' identificazione del modello
Per questa fase utilizzo i dati dal 2014 al 2018 incluso <br>
(ricordarsi che il 2016 fu un anno bisestile)

```python
lenght_tot = conc_max.size
anni_per_validazione = 1

valore_finale_id = lenght_tot - anni_per_validazione*365

yV_id = np.array([item for sublist in zip(*conc_max[1:valore_finale_id]) for item in sublist]) #tolgo l'ultimo anno

u1 = np.array(temp[1:valore_finale_id])
u2 = np.array(vel_vento[1:valore_finale_id])
u3 = np.array(rad_solare[1:valore_finale_id])
u4 = np.array(emissioni[1:valore_finale_id])

u_tot = [u1, u2, u3, u4]

uV_id = np.array(u_tot).transpose()
```

### Variabili per la validazione del modello
Per questa fase utilizzo i dati del 2019 <br>

```python
valore_iniziale_val = lenght_tot - anni_per_validazione*365 

yV_val = np.array([item for sublist in zip(*conc_max[valore_iniziale_val:lenght_tot]) for item in sublist])

u1_val = np.array(temp[valore_iniziale_val:lenght_tot])
u2_val = np.array(vel_vento[valore_iniziale_val:lenght_tot])
u3_val = np.array(rad_solare[valore_iniziale_val:lenght_tot])
u4_val = np.array(emissioni[valore_iniziale_val:lenght_tot])

u_tot_val = [u1_val, u2_val, u3_val, u4_val]

uV_val = np.array(u_tot_val).transpose()
```

## Linear Regression Model

```python
from sklearn.linear_model import LinearRegression

models_lr = []

for ordine_ar in range(1,6):
    
    for ordine_ex in range(0,6):
        
        #per l'ordine della parte autoregressiva e quella esogena ho preso i valori dal codice di Davide
        ordine_parte_autoregressiva = ordine_ar
        ordine_parte_esogena = ordine_ex
        
        ar_ord = ordine_parte_autoregressiva
        ex_ord = np.array(np.ones(4)*ordine_parte_esogena, dtype='i')
        ex_del = np.array(np.zeros(4), dtype = 'i')
        
        #shift dell' input e output per l'indentificazione
        y_id,X_id=dataset_shift(yV_id,uV_id,ar_ord,ex_ord,ex_del)
        
        #shift dell' input e output per la validazione
        y_val,X_val=dataset_shift(yV_val,uV_val,ar_ord,ex_ord,ex_del)
        
        #eliminazione dei valori NaN
        y_id_clean, X_id_clean=clean(y_id,X_id)
        y_val_clean, X_val_clean=clean(y_val,X_val)
        
        # Linear Regression model
        model=LinearRegression()
        
        model.fit(X_id_clean,y_id_clean)
        
        # compio le predizioni con il modello appena calcolato
        ypred_id_clean = model.predict(X_id_clean) # y predicted for 2014-2018 ... come mai lo calcoliamo?
        ypred_val_clean = model.predict(X_val_clean) # y predicted for 2019
        
        # calcolo i diversi errori sui valori appena stimati
        perf_id = valid_perf(y_id_clean,ypred_id_clean)
        perf_val = valid_perf(y_val_clean,ypred_val_clean)

        #calcolo lo score del modello
        score = model.score(X_val_clean, y_val_clean)

        # va a stimare i valori mancanti (NaN)
        y_pred_val = nanpredict(model,X_val)
        y_pred_id = nanpredict(model,X_id)
        
        model_lr = dict(name='LR', model=model, score=score, perf_id=perf_id, perf_val=perf_val, 
                        ar_ord=ar_ord, ex_ord=ex_ord, ex_del=ex_del, X_id=X_id, X_id_clean=X_id_clean, 
                        y_id=y_id, y_id_clean=y_id_clean, X_val=X_val, X_val_clean=X_val_clean, y_val=y_val, 
                        y_val_clean=y_val_clean, y_pred_id=y_pred_id, y_pred_val=y_pred_val)
        
        models_lr.append(model_lr)
```

## Decision Tree Regressor Model

```python
from sklearn.tree import DecisionTreeRegressor

models_dtr = []

for ordine_ar in range(1,6):
    
    for ordine_ex in range(0,6):
             
        #per l'ordine della parte autoregressiva e quella esogena ho preso i valori dal codice di Davide
        ordine_parte_autoregressiva = ordine_ar
        ordine_parte_esogena = ordine_ex
        
        ar_ord = ordine_parte_autoregressiva
        ex_ord = np.array(np.ones(4)*ordine_parte_esogena, dtype='i')
        ex_del = np.array(np.zeros(4), dtype = 'i')
        
        #shift dell' input e output per l'indentificazione
        y_id,X_id=dataset_shift(yV_id,uV_id,ar_ord,ex_ord,ex_del)

        #shift dell' input e output per la validazione
        y_val,X_val=dataset_shift(yV_val,uV_val,ar_ord,ex_ord,ex_del)
        
        #eliminazione dei valori NaN
        y_id_clean, X_id_clean=clean(y_id,X_id)
        y_val_clean, X_val_clean=clean(y_val,X_val)

        model = DecisionTreeRegressor(max_depth=5) #5 è il valore migliore

        model.fit(X_id_clean, y_id_clean)

        # compio le predizioni con il modello appena calcolato
        ypred_id_clean = model.predict(X_id_clean)
        ypred_val_clean = model.predict(X_val_clean)

        # calcolo i diversi errori sui valori appena stimati
        perf_id = valid_perf(y_id_clean,ypred_id_clean)
        perf_val = valid_perf(y_val_clean,ypred_val_clean)

            
        # va a stimare i valori mancanti (NaN)
        y_pred_val = nanpredict(model,X_val)
        y_pred_id = nanpredict(model,X_id)
        
        #calcolo lo score del modello
        score = model.score(X_val_clean, y_val_clean)

        model_dtr = dict(name='DTR', model=model, score=score, perf_id=perf_id, perf_val=perf_val, 
                         ar_ord=ar_ord, ex_ord=ex_ord, ex_del=ex_del, X_id=X_id, X_id_clean=X_id_clean, 
                         y_id=y_id, y_id_clean=y_id_clean, X_val=X_val, X_val_clean=X_val_clean, y_val=y_val, 
                         y_val_clean=y_val_clean, y_pred_id=y_pred_id, y_pred_val=y_pred_val)

        models_dtr.append(model_dtr)
```
